```python
import finviz
from finviz.screener import Screener
import nest_asyncio #to workaround some errors
nest_asyncio.apply()
import pandas as pd
```

```python
filters = ['sh_avgvol_o1000','sh_price_10to50','sh_relvol_o1.5','ta_averagetruerange_o0.5']  #parameters taken from the url https://finviz.com/screener.ashx?v=141&f=sh_avgvol_o1000,sh_price_10to50,sh_relvol_o1.5,ta_averagetruerange_o0.5&ft=3
stock_list = Screener(filters=filters, table='Performance', order='price')  
```

```python
stock_list.to_csv('stock_list.csv')
df=pd.read_csv('stock_list.csv')
df
```
