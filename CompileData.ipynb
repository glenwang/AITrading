```python
#data sources, etc
import pandas as pd
from alpha_vantage.timeseries import TimeSeries
from alpha_vantage.techindicators import TechIndicators
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import time

#screener
import finviz
from finviz.screener import Screener
import nest_asyncio #to workaround some errors
nest_asyncio.apply()
```

```python
tickers=pd.read_csv('stock_list.csv')['Ticker']
tickers
```

```python
# stock data API alpha_vantage, documentation https://www.alphavantage.co/documentation/
key = 'A6QPO3VBTX5UHOAU'
# Chose your output format, or default to JSON (python dict)
ts = TimeSeries(key, output_format='pandas')
ti = TechIndicators(key)


def compileTickersData(tickers):
    compiledData=pd.DataFrame()
    for ticker in tickers:
        data, meta_data = ts.get_intraday(symbol=ticker) #for intraday price, ts.get_intraday. Daily price, ts.get_daily
        data['ticker']=ticker
        compiledData=compiledData.append(data)
        time.sleep(12)
    compiledData=compiledData.rename(columns={'1. open':'open','2. high':'high','3. low':'low', '4. close':'close', '5. volume':'volume'})
    compiledData.to_csv('compiledData.csv')
    return compiledData


cData=compileTickersData(tickers)
cData
```
